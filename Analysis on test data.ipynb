## Compute Accuracy

```python
# load original test data
df = pd.read_csv('data/test.tsv', sep='\t', header=None)
```

```python
df.head()
```

```python
# load BiMPM model predictions with only word embeddings on test data
df_1 = pd.read_csv('~/Downloads/hpc/BiMPM_keras/word/subm_BiMPM_rnn_gru_seq_40_context_100_mp_10_aggreg_300_highway_1_shareRNN_1_drop_0.10              _char_0_k_1_trial_5_db_0-2017-12-16-23-37.csv')
```

```python
df_1.head()
```

```python
def predict_label(df):
    '''predict labels for test data'''
    df['label'] = ''
    for i in range(len(df)):
        if df['is_duplicate'][i] >= 0.5:
            df['label'][i] = 1
        else:
            df['label'][i] = 0
```

```python
def compute_accuracy(df, df_pred):
    '''compute prediction accuracy on test data'''
    cor = 0
    for i in range(len(df[0])):
        if df[0][i] == df_pred['label'][i]:
            cor += 1
    return cor / len(df)
```

```python
predict_label(df_1)
compute_accuracy(df, df_1)
```

```python
# load BiMPM model predictions with both word and character embeddings on test data
df_2 = pd.read_csv('~/Downloads/hpc/BiMPM_keras/result2/subm_BiMPM_rnn_gru_seq_40_context_100_mp_10_aggreg_300_highway_1_shareRNN_1_drop_0.10              _char_1_k_1_trial_5_db_0-2017-12-16-05-27.csv')
```

```python
predict_label(df_2)
compute_accuracy(df, df_2)
```

## Error Analysis

```python
# combine the three dataframes: the original test data, the BiMPM model and the DECATT model predictions (both with character embeddings)
df_combined = pd.DataFrame()
df_combined['test_id'] = df[3]
df_combined['q1'] = df[1]
df_combined['q2'] = df[2]
df_combined['true_label'] = df[0]
df_combined['bimpm_pred'] = df_2['label']
df_combined['decatt_pred'] = df_4['label']
df_combined.head(20)
```

```python
# examples where DECATT correctly predicted but BiMPM incorrectly predicted
df_combined[(df_combined.bimpm_pred != df_combined.true_label) & (df_combined.decatt_pred == df_combined.true_label)].to_csv('compare_1.csv', index=False)
```

```python
# examples where BiMPM correctly predicted but DECATT incorrectly predicted
df_combined[(df_combined.bimpm_pred == df_combined.true_label) & (df_combined.decatt_pred != df_combined.true_label)].to_csv('compare_2.csv', index=False)
```

```python
# examples where both DECATT and BiMPM incorrectly predicted
df_combined[(df_combined.bimpm_pred != df_combined.true_label) & (df_combined.decatt_pred != df_combined.true_label)].to_csv('compare_3.csv', index=False)
```

```python
# examples where both DECATT and BiMPM correctly predicted
df_combined[(df_combined.bimpm_pred == df_combined.true_label) & (df_combined.decatt_pred == df_combined.true_label)].to_csv('compare_4.csv', index=False)
```
